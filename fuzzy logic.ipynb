```python
import numpy as np
# Fuzzy sets A and B
A = {"a": 0.6, "b": 0.8, "c": 0.2, "d": 0.9, "e": 0.5}
B = {"a": 0.2, "b": 1, "c": 0.3, "d": 0.4, "e": 0.1}
# Function to perform fuzzy set union operation
def fuzzy_set_union(A, B):
 y = {}
 for j, k in zip(A, B):
    A_value = A[j]
    B_value = B[k]
    if A_value > B_value:
      y[j] = A_value
    else:
      y[k] = B_value
 return y
# Function to perform fuzzy set intersection operation
def fuzzy_set_intersection(A, B):
 y = {}
 for j, k in zip(A, B):
    A_value = A[j]
    B_value = B[k]
    if A_value < B_value:
      y[j] = A_value
    else:
      y[k] = B_value
 return y
# Function to perform fuzzy set subtraction operation
def fuzzy_set_subtraction(A, B):
 y = {}
 for j, k in zip(A, B):
      A_value = A[j]
      B_value = B[k]
      if A_value > B_value:
        y[j] = A_value - B_value
      else:
        y[k] = 0
 return y
# Function to perform fuzzy set complement operation
def fuzzy_set_complement(A):
 y = {}
 for j in A:
    y[j] = 1 - A[j]
 return y
print("Fuzzy set A:", A)

print("Fuzzy set B:", B)
# Perform fuzzy set union operation
y = fuzzy_set_union(A, B)
print("Fuzzy set union of A and B:", y)
# Perform fuzzy set intersection operation
y = fuzzy_set_intersection(A, B)
print("Fuzzy set intersection of A and B:", y)
# Perform fuzzy set subtraction operation
y = fuzzy_set_subtraction(A, B)
print("Fuzzy set subtraction of A and B:", y)
# Perform fuzzy set complement operation
y = fuzzy_set_complement(A)
print("Fuzzy set complement of A:", y)
```
